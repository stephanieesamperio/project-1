import pandas as pd

# Importing Dataframe 

```python
df = pd.read_csv("shopping_behavior_updated.csv")
```

```python
df
```

#  PD.CUT

```python
df['Generation'] = pd.cut(x=df['Age'], bins=[1,10,26,42,58,68], labels=['Gen Alpha', 'Gen Z', 'Millennial','Gen X','Boomers'])
df
```

# Total Purchase Rank

```python
purchase_rank = df.groupby("Generation")['Age'].count()

purchase_rank
```

# Visuals Bar Graph

```python
Generation_counts = df['Generation'].value_counts()


Generation_counts.plot(kind='bar', rot=0, xlabel='Generation', ylabel='Purchases', title='Count of purchases for each Generation')
```

# More Visuals Pie Chart

```python
Generation_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6, 6), title='Purchases Count')
```

# Finding The Mean 

```python
purchase_rank_mean = df.groupby("Generation")['Age'].mean()

purchase_rank_mean
```
